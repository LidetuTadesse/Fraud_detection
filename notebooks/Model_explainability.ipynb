```python
# Imports
import shap
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create SHAP Explainer for Random Forest
# Use TreeExplainer for Random Forest
rf_model = joblib.load("../models/rf_model.pkl")
X_test = np.load("../data/model_data/X_test.npy")
X_train = np.load("../data/model_data/X_train_sm.npy")
explainer = shap.TreeExplainer(rf_model)
print(X_test.shape)
print(rf_model)
print(rf_model.n_features_in_)
print(X_train.shape)
```

```python
# shap_values = explainer.shap_values(X_test)
X_sample = X_test[:100] 
shap_values = explainer.shap_values(X_sample)
```

```python
# Summary Plot (Global Feature Importance)
# Plot global feature importance

shap.summary_plot(np.mean(np.abs(shap_values), axis=2), X_sample, plot_type="bar")

```

```python
# SHAP Beeswarm Summary Plot
# SHAP summary plot (detailed distribution)
shap.summary_plot(shap_values[:,:, 1], X_sample)
```
